```python
import os, pathlib, glob, shutil

for i in ['test', 'train', 'valid']:
    label_dir = os.path.join('yolo_text_detection.v1i.yolov8-obb/', i, 'labels')
    labels_file = os.listdir(label_dir)
    for label in labels_file:
        with open(os.path.join(label_dir, label), 'r') as f:
            lines = f.readlines()
            # print(lines)
            new_content = ''
            for line in lines:
                new_data = '1'
                for coor in line.split(' ')[1:]:
                    new_data += ' ' +coor
                new_data+='\n'
                new_content+=new_data
            print(new_content)
            f.close()
        with open(os.path.join(label_dir, label), 'w') as f:
            f.write(new_content)
            print('Modification succeed of file {}'.format(label))
```
